```python
y = df.Survived
x = df.drop(['PassengerId', 'Survived', 'Name', 'Ticket'], axis=1)
```

```python
# train, test data split
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123, stratify=y)
```

```python
# load datasets
```

```python
df = pd.read_csv('./ml_source_files/loan_data.csv')
y = df['not.fully.paid']
x = df.drop('not.fully.paid', axis=1)
```

```python
x = pd.get_dummies(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123, stratify=y)
```

```python
# VotingClassifier
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

from sklearn.ensemble import VotingClassifier
```

```python
tree = DecisionTreeClassifier()
log = LogisticRegression()
knn = KNeighborsClassifier()
nb = GaussianNB()
svc = SVC(probability=True)

vot_h = VotingClassifier(estimators = [('log', log), ('tree', tree), ('knn', knn), ('nb', nb), ('svc', svc)], voting='hard')
vot_s = VotingClassifier(estimators = [('log', log), ('tree', tree), ('knn', knn), ('nb', nb), ('svc', svc)], voting='soft')
```

```python
models = [tree, log, knn, nb, svc, vot_h, vot_s]
```

```python
for m in models:
    m.fit(x_train, y_train)
    accuracy = m.score(x_test, y_test)
    
    print(m.__class__.__name__, ':', accuracy)
    print('---------------------')
```
