```python
import mglearn
```

```python
x, y = mglearn.datasets.load_extended_boston()
```

```python
x[:5]
```

```python
y[:5]
```

```python
# y = wx + b
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=111)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)

print('train acc:', lr.score(x_train, y_train))   # 더 높게 나온다..
print('test acc:', lr.score(x_test, y_test) )

# train data의 정확도와 test data의 정확도의 차이가 크다... => 과적합의 가능성 높음..
```

```python
# Ridge, Lasso Elasticnet
```

```python
from sklearn.linear_model import Ridge, Lasso, ElasticNet
```

```python
rid = Ridge(alpha=0.04)
rid.fit(x_train, y_train)

print('train acc:', rid.score(x_train, y_train))  
print('test acc:', rid.score(x_test, y_test)) # 최적화
```

```python
rid = Ridge(alpha=10)
rid.fit(x_train, y_train)

print('train acc:', rid.score(x_train, y_train))  
print('test acc:', rid.score(x_test, y_test))  # 과소적합..
```

```python
ridge_scores = []
for i in np.arange(0.01, 0.2, 0.01):
    rid = Ridge(alpha=i)
    rid.fit(x_train, y_train)
    ridge_scores.append(rid.score(x_test, y_test))
    
plt.plot(np.arange(0.01, 0.2, 0.01), ridge_scores)
```

```python
lasso_scores = []
for i in np.arange(0.0001, 0.01, 0.0001):
    las = Lasso(alpha=i)
    las.fit(x_train, y_train)
    lasso_scores.append(las.score(x_test, y_test))
    
plt.plot(np.arange(0.0001, 0.01, 0.0001), lasso_scores)
plt.show()
```
