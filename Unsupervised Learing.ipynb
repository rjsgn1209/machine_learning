```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# kmeans 
```

```python
from sklearn.datasets import make_blobs

x, y = make_blobs(n_samples = 150,
                 n_features = 2,
                 centers = 3,
                 cluster_std = 0.5,
                 shuffle = True,
                 random_state = 0)
```

```python
x
```

```python
y
```

```python
plt.scatter(x[:,0], x[:,1], c='white', marker='o', edgecolor='black', s=50)
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 3,
           init = 'random',
           max_iter = 300,
           random_state = 111)

y_km = km.fit_predict(x)
y_km
```

```python
km.cluster_centers_
```

```python
plt.figure(figsize=(10, 7))

plt.scatter(x[y_km == 0, 0], x[y_km == 0, 1],
           s = 50, c = 'lightgreen', marker='s', edgecolor='black', label='cls 1')

plt.scatter(x[y_km == 1, 0], x[y_km ==1, 1],
           s = 50, c = 'orange', marker='o', edgecolor='black', label='cls 2')

plt.scatter(x[y_km == 2, 0], x[y_km == 2, 1],
           s = 50, c = 'lightblue', marker='v', edgecolor='black', label='cls 3')

plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
           s = 250, c = 'red', marker='*', edgecolor='black', label='center point')

plt.legend(scatterpoints = 1, loc = 3)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
km.inertia_

# cluster 내부의 std값..
# 작은 값일수록 데이터가 클러스터 내에서 뭉쳐져 있음 => 효율적인 clustering
# 임시적 지표...
```

```python
# 최적의 cluster 갯수 찾기..
```

```python
distortions = []
for i in range(1, 11):
    km = KMeans(n_clusters = i,
               init = 'random',
               random_state = 111)
    km.fit(x)
    distortions.append(km.inertia_)
    
plt.plot(range(1, 11), distortions, marker='o')
plt.show()
```

```python
# silhouette
# 군집 내의 데이터들이 얼마나 결합력있게 그룹핑이 되어있는지..(뭉쳐있는지)
# 군집내의 샘플 간의 거리 - a
# 군집 간의 샘플간의 거리 계산.. - b      
# 군집 내의 데이터는 가까울수록 좋고, 군집 간의 데이터의 거리는 멀수록 좋다..
# silhouette = (b-a)/max(a,b)
```

```python
from matplotlib import cm
from sklearn.metrics import silhouette_samples
```

```python
km = KMeans(n_clusters=3,
           init='k-means++',
           random_state=111)

y_km = km.fit_predict(x)

silhouette_vals = silhouette_samples(x, y_km, metric='euclidean')
```

```python
# silhouette이 가장 정확한 cluster 평가 지표..
# but cluster의 절대적인 평가지표는 존재하지 않는다.
# 시각화를 통한 silhouette 평가
# cluster 별로 silhouette 점수가 고른지 여부 판단 .. - cluster의 쏠림 여부
# silhoueete 평균점수값

y_ax_lower, y_ax_upper = 0, 0
# y_ticks = []
cluster_labels = np.unique(y_km)
n_cluster = cluster_labels.shape[0]

clist = ['g', 'b', 'y']

for i,c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km == c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals)
    plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0,
            edgecolor='none', color=clist[i])
    # yticks.append((y_ax_lower+y_ax_upper)/2)
    y_ax_lower += len(c_silhouette_vals)
    
sihouette_avg = np.mean(silhouette_vals)
plt.axvline(sihouette_avg, color='red', linestyle='--')

plt.show()
```

```python
silhouette_list = []
for i in range(2, 11):
    km = KMeans(n_clusters=i,
               init='k-means++',
               n_init=10,
               max_iter=300,
               random_state=111)
    y_km = km.fit_predict(x)
    silhouette_vals = silhouette_samples(x, y_km, metric='euclidean')
    silhouette_avg = np.mean(silhouette_vals)
    silhouette_list.append(silhouette_avg)
    
plt.plot(range(2, 11), silhouette_list)
```

### Hierachiacl Clustering
* 모든 샘플들의 거리행렬을 계산
* 가장 멀리있는 개체들의 거리에 기반하여 가장 가까운 두개의 군집을 합한다
* 거리행렬을 다시 계산한다
* 하나의 군집이 남을 때 까지 반복한다..
* dendrogram을 드릴 수 있다..

```python
### DBSCAN (밀도기반 클러스터링)
* 밀도 - 특정 반지름(e) 내에 점(데이터)들의 수
* 특정 데이터를 중심으로 반지름(e) 내에 이웃점의 수가 설정한 수 (Minpts)보다 많으면 중심
* 경계점 - e 내에서 중심점은 아니지만 다른 중심점의 e범위 내에 존재하는 경우
* 잡음점 - 중심점도 경계점도 아닌 다른 모든 점..
* 중심점이 서로 다른 군집의 일부가 되면 그군집은 서로 연결되어 하나의 군집이 된다..
* 잡음의 제거가 쉽다,
* 차원이 증가하면 부정적 영향.
```

```python
from sklearn.datasets import make_moons

x, y = make_moons(n_samples=200, noise=0.05, random_state=111)
plt.scatter(x[:, 0], x[:, 1])
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics.cluster import silhouette_score

db = DBSCAN(eps=0.2, min_samples=5, metric='euclidean')
y_db = db.fit_predict(x)
y_db
```

```python
silhouette_score(x, y_db)
```

```python
km = KMeans(n_clusters=3,
          init='random',
          max_iter = 300,
          random_state=111)

y_km = km.fit_predict(x)
```

```python
agg = AgglomerativeClustering(n_clusters = 3)
y_agg = agg.fit_predict(x)

y_agg
```

```python
print('km_sil_score :', np.mean(silhouette_score(x, y_km, metric='euclidean')))
print('db_sil_score :', np.mean(silhouette_score(x, y_db, metric='euclidean')))
print('agg_sil_score :', np.mean(silhouette_score(x, y_agg, metric='euclidean')))
```
